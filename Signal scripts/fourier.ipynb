## Fourier representation of square wave harmonics

```python
import numpy as np
import matplotlib.pyplot as plt
import math

fs = 10000  # Sampling frequency
ts = 1 / fs
t = np.arange(-1, 1, ts)

f = 2  # 2 Hz
omega0 = 2 * np.pi * f
N_values = [n for n in range(100, 1000, 100)]  

fig, ax = plt.subplots(math.ceil(len(N_values)/2), 2, figsize=(12, 8))  # 3x2 grid for 6 plots

def a(k):
    if k == 0:
        return 1 / 2
    else:
        return np.sin(np.pi * k / 2) / (np.pi * k)  # T = 4*T_1



for i, N in enumerate(N_values):
    xt = np.zeros(len(t))  # Reset signal for each N
    for k in range(-N, N + 1):
        xt += a(k) * np.cos(k * omega0 * t)  # Using real-valued cosine terms

    row = i // 2
    col = i % 2
    ax[row, col].plot(t, xt)
    ax[row, col].grid(True)
    ax[row, col].set_xlabel('Time (s)')
    ax[row, col].set_ylabel('Amplitude')
    ax[row, col].set_title(f'Fourier Series Approximation, N={N}')

plt.tight_layout()
plt.show()
```

Frequency domain representation of square wave harmonics

```python
import numpy as np
import matplotlib.pyplot as plt

N = 50
n_values = [4, 8, 12, 16]

fig, ax = plt.subplots(4, 1, figsize=(20, 10))  # 2x2 grid for 4 plots

for i, n in enumerate(n_values):
    A1 = []
    for k in range(-N, N + 1):
        if k == 0:
            A1.append(1 / 2)
        else:
            A1.append((1 / (k * np.pi)) * np.sin(k * 2 * np.pi / n))
    
    k = np.arange(-N, N + 1)
    row = i   # Determine row index
    col = i % 2   # Determine column index

    ax[row].spines["left"].set_position("zero")
    ax[row].spines["bottom"].set_position("zero")
    ax[row].stem(k, A1, basefmt=" ")
    ax[row].set_xlabel("k")
    ax[row].set_ylabel("a_k")
    ax[row].set_title(f"Fourier Coefficients for n={n}")
    ax[row].grid(True)

plt.tight_layout()
plt.show()
```
