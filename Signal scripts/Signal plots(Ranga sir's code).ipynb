DT aperiodic sinusoids

```python
import numpy as np
import matplotlib.pyplot as plt
#import seaborn as sns


fig, ax = plt.subplots(4, figsize=(20,20))
plt.subplots_adjust(left=None, bottom=None, right=None, top=None, wspace=None, hspace=0.5)
A = 1
phi = 0
nmax = 8
n = np.arange(-4*nmax, 4*nmax + 1, 1)
for i, (N, s) in enumerate(zip(list([6, 12, 31/4, (3*np.pi)]), list(['2\pi/6', '2\pi/12', '2\pi/(31/4)', '2/3']))):
    xn = A*np.cos(2*np.pi/N*n + phi)
    ax[i].stem(n, xn, linefmt='-', markerfmt='o', label='$A\cos(\omega_0n + \phi)$, $\phi = 0$')
    ax[i].set_title('$\cos({} \cdot n)$'.format(s))


for i, N in enumerate(list([6, 12, 31/4])):
    xn = A*np.cos(2*np.pi/N*n + phi)
    for j, n_ in enumerate(n):
        if n_%N == 0:
            ax[i].stem(n_, xn[j], linefmt='-r', markerfmt='or')
```

DT sinusoids of different freq.

$x|n| = A cos (\omega _0n+\phi))$

Note that 
1. Only $N=8$ distinct freq. possible in general 

$\omega_0 = 0, 2\pi/8, 4\pi/8, 6\pi,8,10\pi.8,12\pi/8, 14\pi/8$

2. Within these $\omega_0 = 8\pi/8 = \pi$ is the highest frequency.

3. For the case $cos(\omega_0n+\phi)$, there are identical signals even within these 8.

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt

fig, ax = plt.subplots(3,3, figsize=(15,9))
plt.subplots_adjust(left=None, bottom=None, right=None, top=None, wspace=None, hspace=0.5)
A = 1
N = 8
phi = 0
n = np.arange(-4*N, 4*N + 1, 1)
for k in range(0, N+1):
    xn = A*np.cos(k*2*np.pi/N*n + phi)
    ax[k//3, k%3].stem(n, xn, 'b', markerfmt='bo', label='$A\cos(\omega_0n + \phi)$, $\phi = 0$')
    ax[k//3, k%3].set_title('$\cos({0} \cdot 2\pi/{1} \cdot n)$'.format(k, N))

plt.show()
```

CT complex exponentials

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt

fig, ax = plt.subplots(1,2, figsize=(16,4))

t = np.arange(-3*np.pi, 3*np.pi + np.pi/100, np.pi/100)

C = 1
omega0 = 1
phi = 0
r = 0.2
xt1 = np.abs(C)*np.exp(r*t)*np.cos(omega0*t + phi)
ax[0].plot(t/np.pi, xt1, label='$A\cos(\omega_0t + \phi)$, $\omega_0 = {0}$, $\phi = {1}\pi$'.format(omega0, phi/np.pi))
ax[0].plot(t/np.pi, np.exp(r*t), 'r--')
ax[0].plot(t/np.pi, -np.exp(r*t), 'r--')
ax[0].set_ylabel('$x(t) = |C|e^{rt}\cos(\omega_0 t+ \phi)$, $r = 0.2$', loc='top', rotation=0)

C = 1
omega0 = 1
phi = 0
r = -0.2
xt2 = np.abs(C)*np.exp(r*t)*np.cos(omega0*t + phi)
ax[1].plot(t/np.pi, xt2, label='$A\cos(\omega_0t + \phi)$, $\omega_0 = {0}$, $\phi = {1}\pi$'.format(omega0, phi/np.pi))
ax[1].plot(t/np.pi, np.exp(r*t), 'r--')
ax[1].plot(t/np.pi, -np.exp(r*t), 'r--')
ax[1].set_ylabel('$x(t) = |C|e^{rt}\cos(\omega_0 t+ \phi)$, $r = -0.2$', loc='top', rotation=0)

for i in range(0,2):
    ax[i].spines['left'].set_position('zero')
    ax[i].spines['right'].set_color('none')
    ax[i].spines['bottom'].set_position('zero')
    ax[i].spines['top'].set_color('none')
    ax[i].set_xlabel('$t/\pi$', loc='right')
    ax[i].xaxis.set_label_coords(1.05, 0.5)
    ax[i].yaxis.set_label_coords(0.5, 1.05)
    ax[i].grid(True)

plt.subplots_adjust(left=None, bottom=None, right=None, top=None, wspace=None, hspace=0.5)  
plt.show()
```

DT Complex Exponentials: Plotting the Real Part
$|C||\alpha|^n cos(\omega _0*n + \theta)$

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt

fig, ax = plt.subplots(1,2, figsize=(16,4))

n = np.arange(-20, 40 + 1, 1)
C = 1
omega0 = 2*np.pi/16
theta = 0
alpha = 0.889
xn1 = np.abs(C)*(np.abs(alpha))**n*np.cos(omega0*n + theta)
ax[0].stem(n, xn1, label='')
ax[0].plot(n, alpha**n, 'r--')
ax[0].plot(n, -alpha**n, 'r--')
ax[0].set_ylabel('$x[n] = |C||\\alpha|^n\cos(\omega_0 n+ \\theta)$, $\\alpha = 1.08$', loc='top', rotation=0)


n = np.arange(-40, 20 + 1, 1)

alpha = 0.92
xn2 = np.abs(C)*(np.abs(alpha))**n*np.cos(omega0*n + theta)
ax[1].stem(n, xn2, label='')
ax[1].plot(n, alpha**n, 'r--')
ax[1].plot(n, -alpha**n, 'r--')
ax[1].set_ylabel('$x[n] = |C||\\alpha|^n\cos(\omega_0 n+ \\theta)$, $\\alpha = 0.92$', loc='top', rotation=0)

for i in range(0,2):
    ax[i].spines['left'].set_position('zero')
    ax[i].spines['right'].set_color('none')
    ax[i].spines['bottom'].set_position('zero')
    ax[i].spines['top'].set_color('none')
    ax[i].set_xlabel('$n$', loc='right')
    ax[i].xaxis.set_label_coords(1.05, 0.5)
    ax[i].yaxis.set_label_coords(0.5, 1.05)
    ax[i].grid(True)

plt.subplots_adjust(left=None, bottom=None, right=None, top=None, wspace=None, hspace=0.5)  
plt.show()
```

Unit step and impulse functions


```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.ticker import MaxNLocator

fig, ax = plt.subplots(1,2, figsize=(16,4))

n = np.arange(-10, 10 + 1, 1)
u = np.zeros(len(n))
u[n>=0] = 1
ax[0].stem(n, u)
ax[0].set_ylabel('$u[n]$', loc='top', rotation=0)

delta = np.zeros(len(n))
delta[n==0] = 1
ax[1].stem(n, delta)
ax[1].set_ylabel('$\delta[n]$', loc='top', rotation=0)

for i in range(0,2):
    ax[i].spines['left'].set_position('zero')
    ax[i].spines['right'].set_color('none')
    ax[i].spines['bottom'].set_position('zero')
    ax[i].spines['top'].set_color('none')
    ax[i].set_xlabel('$n$', loc='right')
    ax[i].xaxis.set_label_coords(1.05, 0.5)
    ax[i].yaxis.set_label_coords(0.5, 1.05)
    ax[i].xaxis.set_major_locator(MaxNLocator(integer=True))
    ax[i].grid(True)

plt.show()
```
